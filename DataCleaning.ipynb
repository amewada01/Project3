```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
diabetic_df = pd.read_csv("Diabetic_2015.csv", error_bad_lines=False)
```

```python
grocerie_df = pd.read_csv("DataDownload2015.csv")
```

```python
diabetic_df = diabetic_df.dropna(how='any')
```

```python
diabetic_df = diabetic_df[['County', 'State', 'FIPS', 'Percentage']]
diabetic_df['FIPS'] = diabetic_df['FIPS'].astype('int')
diabetic_df.head()
```

```python
columns = grocerie_df.columns

for x in columns:
    print(x)
```

## This Kernel below splits the data between urban vs rural and we will being creating urban data first.

```python
new_grocery = grocerie_df[['CensusTract', 'FIPS', 'State', 'County', 'Urban', 'POP2010', 'OHU2010', 'PovertyRate', 'MedianFamilyIncome', 'TractLOWI', 'TractKids', 'TractSeniors', 'TractWhite', 'TractBlack', 'TractAsian', 'TractNHOPI',
       'TractAIAN', 'TractOMultir', 'TractHispanic', 'TractHUNV', 'LA1and10', 'LAhalfand10', 'LA1and20', 
        'LATracts_half', 'LATracts1', 'LATracts10', 'LATracts20']]
new_grocery = new_grocery[new_grocery.Urban != 0]
```

```python
new_grocery
```

We now have to split the grocery store data into two different data frames. So when we do a groupby some of the values need to be summed together and other values need to be averaged.

Summed:
1. POP2010
2. OHU2010
3. TractLOWI
4. TractKids
5. TractSeniors
6. TractWhite
7. TractBlack
8. TractAsian
9. TractNHOPI
10. TractAIAN
11. TractOMultir
12. TractHispanic
13. TractHUNV

Averaged:
1. PovertyRate
2. MedianFamilyIncome
3. LA1and10
4. LAhalfand10
5. LA1and20
6. LATracts_half
7. LATract1
8. LAtract10
9. LAtract20

```python
new_grocery.columns
```

```python
summed_grocery = new_grocery[['FIPS', 'POP2010', 'OHU2010','TractLOWI', 'TractKids',
       'TractSeniors', 'TractWhite', 'TractBlack', 'TractAsian', 'TractNHOPI',
       'TractAIAN', 'TractOMultir', 'TractHispanic', 'TractHUNV']]
summed_grocery
```

```python
grouped_summed_grocery = summed_grocery.groupby(['FIPS']).sum()
grouped_summed_grocery
```

```python
averaged_grocery = new_grocery[['FIPS','PovertyRate', 'MedianFamilyIncome', 'LA1and10', 'LAhalfand10', 'LA1and20', 'LATracts_half', 'LATracts1', 'LATracts10', 'LATracts20']]
averaged_grocery
```

```python
grouped_averaged_grocery = averaged_grocery.groupby(['FIPS']).mean()
grouped_averaged_grocery
```

```python
merged_grouped_data = pd.merge(grouped_summed_grocery, grouped_averaged_grocery, on="FIPS", how="outer")
merged_grouped_data
```

```python
combined_data = pd.merge(merged_grouped_data, diabetic_df, on='FIPS', how='outer')
```

```python
combined_data
```

```python
combined_data = combined_data.dropna(how='any')
combined_data = combined_data[combined_data.Percentage != 'No Data']
combined_data['Percentage'] = combined_data['Percentage'].astype('float')
percentages = combined_data['Percentage']
new_value = []
for x in percentages:
    new_value.append(round(x/100,3))
combined_data['Percentage'] = new_value
combined_data.head()
```

```python
combined_data.to_csv("UrbanData.csv", index=False, header=True)
```

## From this point onwards we will be filtering the data to only see Rural areas 

```python
rural_new_grocery = grocerie_df[['CensusTract', 'FIPS', 'State', 'County', 'Urban', 'POP2010', 'OHU2010', 'PovertyRate', 'MedianFamilyIncome', 'TractLOWI', 'TractKids', 'TractSeniors', 'TractWhite', 'TractBlack', 'TractAsian', 'TractNHOPI',
       'TractAIAN', 'TractOMultir', 'TractHispanic', 'TractHUNV', 'LA1and10', 'LAhalfand10', 'LA1and20', 
        'LATracts_half', 'LATracts1', 'LATracts10', 'LATracts20']]
rural_new_grocery = rural_new_grocery[rural_new_grocery.Urban != 1]
```

```python
rural_new_grocery
```

```python
rural_summed_grocery = rural_new_grocery[['FIPS', 'POP2010', 'OHU2010','TractLOWI', 'TractKids',
       'TractSeniors', 'TractWhite', 'TractBlack', 'TractAsian', 'TractNHOPI',
       'TractAIAN', 'TractOMultir', 'TractHispanic', 'TractHUNV']]
rural_summed_grocery
```

```python
rural_grouped_summed = rural_summed_grocery.groupby(['FIPS']).sum()
rural_grouped_summed
```

```python
rural_averaged_grocery = rural_new_grocery[['FIPS','PovertyRate', 'MedianFamilyIncome', 'LA1and10', 'LAhalfand10', 'LA1and20', 'LATracts_half', 'LATracts1', 'LATracts10', 'LATracts20']]
rural_averaged_grocery
```

```python
rural_grouped_averaged = rural_averaged_grocery.groupby(['FIPS']).mean()
rural_grouped_averaged
```

```python
rural_merged_grouped_data = pd.merge(rural_grouped_summed, rural_grouped_averaged, on="FIPS", how="outer")
rural_merged_grouped_data
```

```python
rural_combined_data = pd.merge(rural_merged_grouped_data, diabetic_df, on='FIPS', how='outer')
rural_combined_data
```

```python
rural_combined_data = rural_combined_data.dropna(how='any')
rural_combined_data = rural_combined_data[rural_combined_data.Percentage != 'No Data']
rural_combined_data['Percentage'] = rural_combined_data['Percentage'].astype('float')
rural_percentages = rural_combined_data['Percentage']
new_value_rural = []
for x in rural_percentages:
    new_value_rural.append(round(x/100,3))
rural_combined_data['Percentage'] = new_value_rural
rural_combined_data
```

```python
rural_combined_data.to_csv("RuralData.csv", index=False, header=True)
```
